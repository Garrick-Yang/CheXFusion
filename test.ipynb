```python
data = df_train_labeled

# Extract the relevant columns from Adenopathy onwards
disease_columns = data.columns[data.columns.get_loc('Adenopathy'):]

# Group by study_id and sum the values for each disease to get the number of cases per study_id
grouped_data = data.groupby('study_id')[disease_columns].sum()

# Any value greater than 0 means the person has the disease, so we convert to binary
grouped_data = (grouped_data > 0).astype(int)

# Sum across all study_ids to get the number of people with each disease
disease_counts = grouped_data.sum().reset_index()
disease_counts.columns = ['Disease', 'Count']

disease_counts
```
